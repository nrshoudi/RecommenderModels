```python
import json
businesses = []
for line in open('yelp_business.json', 'r'):
    businesses.append(json.loads(line))
```

```python
len(businesses)
```

```python
categories = []
for b in businesses:
    if b['categories'] not in categories:
        categories.append(b['categories'])
categories
```

```python
withcategories = list(filter(lambda b: b['categories'] != None , businesses))
len(withcategories)
```

```python
restaurants = list(filter(lambda b: 'food' in b['categories'].lower() or 'restaurant' in b['categories'].lower(), withcategories))
```

```python
len(restaurants)
```

```python
restaurants[0]
```

```python
restaurant_ids = [r['business_id'] for r in restaurants]
```

```python
reviews = []
for line in open('sub_yelp_reviews.json', 'r'):
    reviews.append(json.loads(line))
```

```python
len(reviews)
```

```python
reviews[0]
```

```python
restaurant_reviews = list(filter(lambda r: r['business_id'] in restaurant_ids, reviews))
```

```python
# from surprise import KNNBasic
# knn = KNNBasic()
from surprise import SVDpp
svdpp = SVDpp()
```

```python
svdpp_cv = cross_validate(svdpp, data, measures=['RMSE', 'MAE'], cv=5)
```

```python
from surprise import NMF

nmf = NMF()
```

```python
nmf_cv = cross_validate(nmf, data, measures=['RMSE', 'MAE'])
```

```python
import matplotlib.pyplot as plt 


x_algo = ['NMF', 'SVD', "SVD++"]
all_algos_cv = [nmf_cv, svd_cv, svdpp_cv]
rmse_cv = [round(res['test_rmse'].mean(), 4) for res in all_algos_cv]
mae_cv = [round(res['test_mae'].mean(), 4) for res in all_algos_cv]

plt.figure(figsize=(20,5))

plt.subplot(1, 2, 1)
plt.title('Comparison of Algorithms on RMSE', loc='center', fontsize=15)
plt.plot(x_algo, rmse_cv, label='RMSE', color='darkgreen', marker='o')
plt.xlabel('Algorithms', fontsize=15)
plt.ylabel('RMSE Value', fontsize=15)
plt.legend()
plt.grid(ls='dashed')
plt.show()
```
